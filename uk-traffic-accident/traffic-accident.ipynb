Aşağıdaki sütunları veri setinden çıkarıyoruz:

```python
data = datas.drop(['Accident_Index', 'Location_Easting_OSGR', 'Location_Northing_OSGR', 
                   'Junction_Detail', 'LSOA_of_Accident_Location'], axis=1)

data = data.dropna(subset=['Longitude', 'Latitude'])

# Time sütunundaki eksikleri "00:00" olarak dolduruyoruz.
data['Time'] = data['Time'].fillna('00:00')
data = data.drop('Junction_Control', axis=1)
data = data.dropna()

# Date sütununu datetime yap
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
```

1. Trafik akışının değişmesi kazaları nasıl etkiler?

Burada trafik akışı ile ilgili verilerimizde hangi sütunlar var ona bakalım.
Number_of_Vehicles, Number_of_Casualties, Accident_Severity gibi sütunlar kaza ile ilgili.

Trafik akışının değişmesi genelde zaman, yol tipi, hız limiti, günün saati gibi değişkenlerle ilişkilidir.

Adım 1: Zaman dilimlerine göre (örneğin gün, saat, yıl) kaza sayıları ve şiddetleri inceleyelim.

```python
# Saat ve gün bilgilerini çıkaralım
data['Hour'] = data['Time'].apply(lambda x: int(x.split(':')[0]) if ':' in x else 0)
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['DayOfWeek'] = data['Date'].dt.dayofweek  # Pazartesi=0

# Saatlere göre kaza sayısı
hourly_accidents = data.groupby('Hour').size()

# Yıllara göre kaza sayısı
yearly_accidents = data.groupby('Year').size()

# Günlere göre kaza sayısı
weekday_accidents = data.groupby('DayOfWeek').size()

print("hourly_accidents:", hourly_accidents)
```

```python
print("yearly_accidents:",yearly_accidents)
```

```python
print("weekday_accidents:",weekday_accidents)
```

Adım 2: Trafik akışını etkileyen faktörler ile kaza ilişkisi
Mesela hız limiti, yol tipi, ışık koşulları gibi.

```python
# Ortalama kaza şiddeti hız limitine göre
severity_by_speed = data.groupby('Speed_limit')['Accident_Severity'].mean()

# Kaza sayısı yol tipine göre
accidents_by_road_type = data.groupby('Road_Type').size()
```

Zaman İçinde Kaza Oranlarını Tahmin Etmek (Regression)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Yıllara göre kaza sayıları
accidents_per_year = data["Year"].value_counts().sort_index()

# X ve y verileri
X = accidents_per_year.index.values.reshape(-1, 1)  # Bağımsız değişken: Yıl
y = accidents_per_year.values                      # Bağımlı değişken: Kaza sayısı
```

```python
# Modeli oluştur ve eğit
model = LinearRegression()
model.fit(X, y)

# Tahmin yap (2017-2025 arası da dahil)
future_years = np.arange(2005, 2026).reshape(-1, 1)
predictions = model.predict(future_years)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='blue', label='Gerçek Veriler')
plt.plot(future_years, predictions, color='red', linestyle='--', label='Tahmin Edilen Trend')

plt.xlabel("Yıl", weight='bold')
plt.ylabel("Kaza Sayısı", weight='bold')
plt.title("Yıllara Göre Trafik Kazaları Tahmini", weight='bold')
plt.legend()
plt.grid(True)
plt.show()
```

Regresyon doğrusu, kaza sayısının zamanla azaldığını gösterecektir.
2015 ve sonrası için de düşüş trendi büyük ihtimalle devam edecek.
Böylece, kaza oranlarını zaman içinde tahmin edebiliriz sonucuna ulaşmış oluruz

4. Kırsal ve Kentsel Alanlar Nasıl Farklılaştı?

Bu soruyu analiz etmek için Urban_or_Rural_Area sütununu kullanacağız.

```python
print(data["Urban_or_Rural_Area"].value_counts(dropna=False))
```

2 → Rural (Kırsal)
1 → Urban (Kentsel)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Etiketleri eşleştir
area_map = {1: "Urban", 2: "Rural"}
data["Area_Label"] = data["Urban_or_Rural_Area"].map(area_map)

# Sayım
area_counts = data["Area_Label"].value_counts()

# Grafik çiz
plt.figure(figsize=(6,4))
sns.barplot(x=area_counts.index, y=area_counts.values, palette="Set2")
plt.title("Kaza Sayıları: Kentsel vs. Kırsal", weight='bold')
plt.xlabel("Alan Türü", weight='bold')
plt.ylabel("Kaza Sayısı", weight='bold')
plt.grid(True)
plt.show()
```

Kırsal alanlarda (Rural) kaza sayısı daha fazla görünüyor.
Bu sonuç, trafik akışı, yol yapısı ya da hız limitleri gibi faktörlerin kırsalda daha fazla kazaya neden olabileceğine işaret edebilir.

Kazaların Şiddeti (Accident Severity) kırsal ve kentsel alanlarda nasıl?

```python
# Şiddet etiketleri
severity_map = {1: "Fatal", 2: "Serious", 3: "Slight"}
data["Severity_Label"] = data["Accident_Severity"].map(severity_map)

# Kırsal ve kentsel alanları etiketlemiştik: "Area_Label"
# Grupla
severity_by_area = data.groupby(["Area_Label", "Severity_Label"]).size().reset_index(name='Count')

# Grafik
plt.figure(figsize=(8,6))
sns.barplot(data=severity_by_area, x="Area_Label", y="Count", hue="Severity_Label", palette="Set2")
plt.title("Kaza Şiddeti: Kentsel vs. Kırsal", weight='bold')
plt.xlabel("Alan Türü", weight='bold')
plt.ylabel("Kaza Sayısı", weight='bold')
plt.legend(title="Kaza Şiddeti")
plt.grid(True)
plt.show()
```

Kaza verilerini “Urban_or_Rural_Area” değişkenine göre incelediğimizde, kırsal (rural) alanlarda kazaların sayısı kentsel (urban) alanlara kıyasla hem genel olarak daha fazla hem de daha şiddetlidir.
Özellikle ölümcül (fatal) ve ciddi (serious) kazaların kırsal bölgelerde çok daha fazla olduğu görülmektedir.
Şehir içi (urban) bölgelerde ise kazaların büyük çoğunluğu hafif (slight) kategorisindedir.

Kırsalda yolların daha hızlı olması (daha yüksek hız limiti),
